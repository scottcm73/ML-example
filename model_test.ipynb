```python


def dummify(df):
    train_df_d = pd.get_dummies(df['Species'])
    print(train_df_d)
    df=df.drop(columns=['Species'])
    df=df.merge(train_df_d, left_index = True, right_index = True )    
    return df

def do_prediction(X_test, model):
    y_pred=model.predict(X_test)
    return y_pred


def get_data
    this_df = pd.read_csv('fish_participant.csv')
    print(this_df.describe())

    df=dummify(this_df)


# Multiple Linear Regression
print(df.columns)

print(df.columns)
X_train = df.iloc[:,df.columns != 'Weight']
y_train = df.iloc[:, 0]

model = LinearRegression()
model.fit(X_train, y_train)

coeff_df = pd.DataFrame(model.coef_, X_train.columns, columns=['Coefficient'])
print(coeff_df)

this_df2 = pd.read_csv('fish_holdout_demo.csv')
print(this_df2.describe())

df2=dummify(this_df2)
#make X_test with Dummy variables
X_test = df2.iloc[:,df2.columns != 'Weight']

print(X_test.shape)

y_test= df2.iloc[:, 0]
y_true = y_test
print(y_true)
y_pred=do_prediction(X_test, model)

y_pred=abs(y_pred)

print(y_pred)

mse=mean_squared_error(y_true, y_pred)
print(mse)

#SVR
n_samples, n_features = 10, 5
rng = np.random.RandomState(5)
y = rng.randn(n_samples)
X = rng.randn(n_samples, n_features)
model2 = SVR(C=1.0, epsilon=.1)
model2.fit(X_train, y_train)
y_pred = model2.predict(X_test)
print(y_pred)
mse = mean_squared_error(y_true, y_pred)
print(mse)

#  Length Average * Height * Width *.27 (Somehow this is the best one!!)

model3=X_test.assign(y_pred=(X_test["Length1"]+ X_test["Length2"]+ X_test["Length3"])/3 * X_test["Height"]*X_test["Width"]*.26)

print(model3)

y_prediction=model3[['y_pred']].to_numpy()
y_pred=y_prediction.ravel()
print(y_pred)

mse3 = mean_squared_error(y_true, y_pred)

# Still quite high??
print(mse3)

```
